# 219. Contains Duplicate II

Given an array of integers and an integer k, find out whether there are two distinct indices i and j in the array such that nums[i] = nums[j] and the absolute difference between i and j is at most k.

## Example 1:

    Input: nums = [1,2,3,1], k = 3
    Output: true

## Example 2:
    Input: nums = [1,0,1,1], k = 1
    Output: true

## Example 3:

    Input: nums = [1,2,3,1,2,3], k = 2
    Output: false

```
# O(n^2)
class Solution:
    def containsNearbyDuplicate(self, nums, k):
        n = len(nums)
        for i in range(n):
            m = min(n,i+k+1)
            print(nums[i:m])
            for j in range(i+1,m):
                if nums[i] == nums[j]:
                    return True
        return False
```

```
# O(n)
class Solution:
    def containsNearbyDuplicate(self, nums, k):
        count = {}
        for i in range(len(nums)):

            n = nums[i]
            if n not in count:
                count[n] = i
            else:
                if (i - count[n]) <=k:
                    return True
                else:
                    count[n] = i

        return False
```

```
assert(Solution().containsNearbyDuplicate([1,2,3,1],3) == True)
assert(Solution().containsNearbyDuplicate([1,0,1,1],1) == True)
assert(Solution().containsNearbyDuplicate([1,2,3,1,2,3],2) == False)
```

```
nums = [i for i in range(-24500,30000,1)]
assert(Solution().containsNearbyDuplicate(nums,35000) == False)
```
