# 206. Reverse Linked List

## Example:

    Input: 1->2->3->4->5->NULL
    Output: 5->4->3->2->1->NULL

Follow up:

A linked list can be reversed either iteratively or recursively. Could you implement both?

```
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def reverseList(self, head: ListNode) -> ListNode:
        if not head:
            return None
        tail = head
        new_head = head
        h = tail.next
        # 1(tail,new_head),2(h),3,4,5
        while h:
            # 2(h),1(tail,new_head),3,4,5
            tail.next = h.next
            h.next = new_head
            # 2(h,new_head),1(tail),3,4,5
            new_head = h
            # 2(new_head),1(tail),3(h),4,5
            h = tail.next
        return new_head
```
