# 371. Sum of Two Integers

Calculate the sum of two integers a and b, but you are not allowed to use the operator + and -.

## Example 1:

    Input: a = 1, b = 2
    Output: 3

## Example 2:

    Input: a = -2, b = 3
    Output: 1




```
from typing import *

# 低位a^b ，高位 a&b
class Solution:
    def getSum(self, a: int, b: int) -> int:
        if a == 0:
            return b
        elif b == 0:
            return a
        else:
            while True:
                sum = a^b
                carrier = a&b
                if carrier == 0:
                    break
                a = sum
                b = carrier<<1
        return sum

# assert(Solution().getSum(1,2) == 3)
assert(Solution().getSum(1,3) == 4)
```
