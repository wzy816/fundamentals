# Remove Duplicates from Sorted List

Given a sorted linked list, delete all duplicates such that each element appear only once.

```python
# easy
class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None
        
class Solution:
    def deleteDuplicates(self, head: ListNode) -> ListNode:
        p = head
        while p and p.next :
            q = p.next
            if q.val == p.val:
                p.next = q.next
            else:
                p = q
        return head
```
