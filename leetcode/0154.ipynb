# 154. Find Minimum in Rotated Sorted Array II
Hard

Suppose an array sorted in ascending order is rotated at some pivot unknown to you beforehand.

(i.e.,  [0,1,2,4,5,6,7] might become  [4,5,6,7,0,1,2]).

Find the minimum element.

The array may contain duplicates.

## Example 1:

    Input: [1,3,5]
    Output: 1

## Example 2:

    Input: [2,2,2,0,1]
    Output: 0

## Note:

    This is a follow up problem to Find Minimum in Rotated Sorted Array.
    Would allow duplicates affect the run-time complexity? How and why?



```
from typing import *


class Solution:
    def findMin(self, nums: List[int]) -> int:

        l = len(nums)
        left = 0
        right = l - 1
        # already ascent order
        if nums[left] < nums[right]:
            return nums[left]

        # skip dup from both ends
        while left < l-1 and nums[left] == nums[left+1]:
            left +=1
        while right > 0 and nums[right] == nums[right-1]:
            right -=1

        # make sure nums[left] > nums[right]
        while left + 1 < right:
            mid = (left+right) //2
            if nums[mid] == nums[left]:
                left = mid
            elif nums[mid] < nums[left]:
                right = mid
            else:
                left = mid
        return nums[right]

assert(Solution().findMin([1,3,5]) == 1)     
assert(Solution().findMin([2,2,2,0,1]) == 0)     
assert(Solution().findMin([2,2,2,2]) == 2)
assert(Solution().findMin([3,1,3]) == 1)    
assert(Solution().findMin([10,1,10,10,10]) == 1)  
```
