# Remove Duplicates from Sorted List II
Given a sorted linked list, delete all nodes that have duplicate numbers, leaving only distinct numbers from the original list.

Return the linked list sorted as well.

```python
class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None


class Solution:
    def deleteDuplicates(self, head: ListNode) -> ListNode:
        h = None # output head
        p = None # pointer in output list

        dup = None # store duplicate value

        n = head
        while n:
        
            # if same value, skip and store value
            if n.next and n.val == n.next.val:
                dup = n.val
                
            else:

                if dup is None or (n.next is None and n.val != dup) or (n.val != dup):
                # take current node to output list
                # use current node as output head if output head doesn't exists
                # otherwise append current node to output list
                    print('take',n.val)
                    if h is None:
                        h = n
                        p = h
                    else:
                        p.next = n
                        p = n
                # reset dup
                dup = None

        n = n.next

        if p:
            p.next = None
        return h
```
