# 137. Single Number II

Given a non-empty array of integers, every element appears three times except for one, which appears exactly once. Find that single one.

## Note:

Your algorithm should have a linear runtime complexity. Could you implement **without using extra memory**?

## Example 1:

    Input: [2,2,3,2]
    Output: 3
## Example 2:

    Input: [0,1,0,1,0,1,99]
    Output: 99

```
# using a dict
from typing import *
class Solution:
    def singleNumber(self, nums: List[int]) -> int:
        count= {}
        for n in nums:
            if n in count:
                if count[n] ==1:
                    count[n] =2
                else:
                    del count[n]
            else:
                count[n] = 1
        return list(count.keys())[0]
```

```
# use two counters and bit ops
# refer to https://leetcode.com/problems/single-number-ii/discuss/43295/Detailed-explanation-and-generalization-of-the-bitwise-operation-method-for-single-numbers
from typing import *

class Solution:
    def singleNumber(self, nums: List[int]) -> int:
        one = 0
        two = 0
        for num in nums:
            two ^= (one & num)
            one ^= num
            mask = ~(two & one)
            two &=mask
            one &=mask
        return one
```

```
assert(Solution().singleNumber([2,2,3,2]) == 3)
assert(Solution().singleNumber([0,1,0,1,0,1,99]) == 99)
```
