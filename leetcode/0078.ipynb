# 78. Subsets

Given a set of distinct integers, nums, return all possible subsets (the power set).

Note: The solution set must not contain duplicate subsets.

# Example

    Input: nums = [1,2,3]
    Output:
    [
    [3],
    [1],
    [2],
    [1,2,3],
    [1,3],
    [2,3],
    [1,2],
    []
    ]

```
from typing import * 
class Solution:
    def subsets(self, nums: List[int]) -> List[List[int]]:
        output = [[]]
        def dfs(arr:List[int], rest:List[int], container:List[List[int]]):
            if len(rest) == 0:
                return
            for i in range(len(rest)):
                new_arr = arr + [rest[i]]
                new_rest = rest[i+1:]
                container.append(new_arr)
                dfs(new_arr, new_rest, container)
        dfs([],nums,output)
        return output

assert(sorted(Solution().subsets([1,2,3])) == sorted([[3],[1],[2],[1,2,3],[1,3],[2,3],[1,2],[]]))
```
