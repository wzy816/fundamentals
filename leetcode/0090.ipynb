# 90. Subsets II

Given a collection of integers that might contain duplicates, nums, return all possible subsets (the power set).

Note: The solution set must not contain duplicate subsets.

## Example:

    Input: [1,2,2]
    Output:
    [
    [2],
    [1],
    [1,2,2],
    [2,2],
    [1,2],
    []
    ]

```
from typing import *

class Solution:
    def subsetsWithDup(self, nums: List[int]) -> List[List[int]]:

        nums.sort()

        def dfs(arr,rest,container):
            if len(rest) == 0:
                return
            prev = None
            for i in range(len(rest)):
                if prev == rest[i]:
                    continue
                container.append(arr+[rest[i]])
                dfs(arr+[rest[i]], rest[i+1:], container)
                prev = rest[i]
        output= [[]]
        dfs([],nums,output)
        return output        

assert(sorted(Solution().subsetsWithDup([1,2,2])) == sorted([[2],[1],[1,2,2],[2,2],[1,2],[]]))
assert(sorted(Solution().subsetsWithDup([1,1,2,2])) == sorted([[],[1],[1,1],[1,1,2],[1,1,2,2],[1,2],[1,2,2],[2],[2,2]]))
```
