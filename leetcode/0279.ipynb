# 279. Perfect Squares

Given a positive integer n, find the least number of perfect square numbers (for example, 1, 4, 9, 16, ...) which sum to n.

## Example 1:

    Input: n = 12
    Output: 3 
    Explanation: 12 = 4 + 4 + 4.
## Example 2:

    Input: n = 13
    Output: 2
    Explanation: 13 = 4 + 9.

```
from typing import types
from collections import deque
import math
class Solution:
    def numSquares(self, n: int) -> int:
        cache = set()
        remain = deque([(n,0)])
        squares = [ i * i for i in range(int(math.sqrt(n)+1))]
        while len(remain)> 0:
            
            r, step = remain.popleft()
            for sqr in squares:
                nr = r- sqr
                if nr < 0: 
                    break
                if nr == 0:
                    # print('step',step)
                    return step+1
                if nr > 0 and nr not in cache:
                    remain.append((nr, step+1))
                    cache.add(nr)


assert(Solution().numSquares(13) == 2)
assert(Solution().numSquares(12) == 3)
assert(Solution().numSquares(7168) == 4)
```
