# 40. Combination Sum II

Given a collection of candidate numbers (candidates) and a target number (target), find all unique combinations in candidates where the candidate numbers sums to target.

Each number in candidates may only be used once in the combination.

Note:

All numbers (including target) will be positive integers.
The solution set must not contain duplicate combinations.

## Example 1:

Input: candidates = [10,1,2,7,6,1,5], target = 8,
A solution set is:
[
  [1, 7],
  [1, 2, 5],
  [2, 6],
  [1, 1, 6]
]

## Example 2:

Input: candidates = [2,5,2,1,2], target = 5,
A solution set is:
[
  [1,2,2],
  [5]
]

```
from typing import *
class Solution:
    def combinationSum2(self, candidates: List[int], target: int) -> List[List[int]]:

        candidates.sort()

        def DFS(arr, arr_sum, rest, container):
            if arr_sum > target:
                return
            if arr_sum == target:
                container.append(arr)
                print(container)
            prev = None
            for i in range(len(rest)):
                if prev == rest[i]:
                    continue
                DFS(arr+[rest[i]], arr_sum+rest[i], rest[i+1:], container)
                prev = rest[i]
        
        output = []
        DFS([], 0, candidates, output)
        return output
        
assert(
    sorted(Solution().combinationSum2(  [10,1,2,7,6,1,5], 8))
    == sorted( [ [1, 7], [1, 2, 5], [2, 6], [1, 1, 6] ])
)
assert(
    sorted(Solution().combinationSum2( [2,5,2,1,2], 5))
    == sorted(  [ [1,2,2], [5] ])
)
```
