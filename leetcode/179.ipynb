# Largest Number

Given a list of non negative integers, arrange them such that they form the largest number.

## Example 1:

Input: [10,2]
Output: "210"

## Example 2:

Input: [3,30,34,5,9]
Output: "9534330"

Note: The result may be very large, so you need to return a string instead of an integer.

```
class Solution:
    def largestNumber(self, nums):
        def compare(n1,n2):
            s1 = str(n1)
            s2 = str(n2)
            return int(s1+s2) > int(s2+s1)

        def merge_sort(numbers):
            
            n = len(numbers)
            if n == 1:
                return numbers
            elif n == 2:
                if compare(numbers[0],numbers[1]):
                    return numbers
                else:
                    return [numbers[1],numbers[0]]
            else:
                l = int(n/2)
                left = merge_sort(numbers[0:l])
                right = merge_sort(numbers[l:n])
                ret = []
                i = 0
                j = 0
                while i<len(left) or j < len(right):
                    if j == len(right):
                        ret = ret + left[i:len(left)]
                        break
                    elif i ==len(left):
                        ret = ret + right[j:len(right)]
                        break
                    elif compare(left[i],right[j]):
                        ret.append(left[i])
                        i = i+1
                    else:
                        ret.append(right[j])
                        j = j+1
                return ret

        ns = merge_sort(nums)
        return str(int(''.join([str(item) for item in ns])))    
```

```
assert(Solution().largestNumber([10,2]) == "210")
assert(Solution().largestNumber([3,30,34,5,9]) == "9534330")
assert(Solution().largestNumber([999999998,999999997,999999999]) == "999999999999999998999999997")
```
