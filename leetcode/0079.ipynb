# 79. Word Search

Given a 2D board and a word, find if the word exists in the grid.

The word can be constructed from letters of sequentially adjacent cell, where "adjacent" cells are those horizontally or vertically neighboring. The same letter cell may not be used more than once.

## Example:

    board =
    [
    ['A','B','C','E'],
    ['S','F','C','S'],
    ['A','D','E','E']
    ]

    Given word = "ABCCED", return true.
    Given word = "SEE", return true.
    Given word = "ABCB", return false.
 

## Constraints:

board and word consists only of lowercase and uppercase English letters.
- 1 <= board.length <= 200
- 1 <= board[i].length <= 200
- 1 <= word.length <= 10^3

```
from typing import *

class Solution:
    def exist(self, board: List[List[str]], word: str) -> bool:
        width = len(board[0])
        height = len(board)
        def search (b, i,j, word_left):
            if len(word_left) == 0:
                return True
            ret = False
            candidates = []
            # search left
            if j>=1 and b[i][j-1] is not None:
                candidates.append([i,j-1])
            # search right    
            if j<(width-1) and b[i][j+1] is not None:
                candidates.append([i,j+1])
            # search up
            if i>=1 and b[i-1][j] is not None:
                candidates.append([i-1,j])
            # search right
            if i<(height-1) and b[i+1][j] is not None:
                candidates.append([i+1,j])
            for c in candidates:
                if b[c[0]][c[1]] == word_left[0]:
                    b2 = [ x[:] for x in b]
                    b2[c[0]][c[1]] = None
                    ret = ret or search(b2, c[0], c[1], word_left[1:])
            return ret
        r = False
        for i,row in enumerate(board):
            for j,item in enumerate(row):
                if item == word[0]:
                    b2 = [ x[:] for x in board]
                    b2[i][j] = None
                    r = r or search(b2,i,j,word[1:])
        return r

board =[
['A','B','C','E'],
['S','F','C','S'],
['A','D','E','E']
]
assert(Solution().exist(board, "ABCCED") == True)
assert(Solution().exist(board, "SEE") == True)
assert(Solution().exist(board, "ABCB") == False)
assert(Solution().exist([["a","a"]],"aaa") == False)
```
