# Odd Even Linked List

Given a singly linked list, group all odd nodes together followed by the even nodes. Please note here we are talking about the node number and not the value in the nodes.

You should try to do it in place. The program should run in O(1) space complexity and O(nodes) time complexity.

```python
class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None

# use three pointers
def oddEvenList(head: ListNode) -> ListNode:
    o_tail = head
    e_head = head.next
    e_tail = head.next
    while o_tail and e_tail and e_tail.next:
        o_tail2 = e_tail.next
        e_tail2 = o_tail2.next
        o_tail.next = o_tail2
        o_tail2.next = e_head
        e_tail.next = e_tail2
        o_tail = o_tail2
        e_tail = e_tail2
    return head

n = ListNode(0)
p = n
for i in range(1,10):
    l = ListNode(i)
    p.next = l
    p = l

head = oddEvenList(n)
while head:
    print(head.val)
    head = head.next
```

```python
class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None


class Solution:
    def oddEvenList(self, head: ListNode) -> ListNode:
        if head == None or head.next == None:
            return head
        odd = head
        even_head = head.next
        even = even_head
        t = even.next
        i = 1
        head.next = None
        while t:
            if i %2 ==1:
                odd.next = t
                odd = t
            else:
                even.next = t
                even = t
            t = t.next
            i +=1
        odd.next = even_head
        even.next = None
        return head

n = ListNode(0)
p = n
for i in range(1,5):
    l = ListNode(i)
    p.next = l
    p = l

head = Solution().oddEvenList(n)
while head:
    print(head.val)
    head = head.next
```
