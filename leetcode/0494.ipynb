# 494. Target Sum

You are given a list of non-negative integers, a1, a2, ..., an, and a target, S. Now you have 2 symbols + and -. For each integer, you should choose one from + and - as its new symbol.

Find out how many ways to assign symbols to make sum of integers equal to target S.

## Example 1:

    Input: nums is [1, 1, 1, 1, 1], S is 3. 
    Output: 5
    Explanation: 

    -1+1+1+1+1 = 3
    +1-1+1+1+1 = 3
    +1+1-1+1+1 = 3
    +1+1+1-1+1 = 3
    +1+1+1+1-1 = 3

There are 5 ways to assign symbols to make the sum of nums be target 3.

```
from typing import *

# dfs
class Solution:
    def findTargetSumWays(self, nums: List[int], S: int) -> int:
        self.total = 0
        def find(sum, cur):
            if cur == len(nums):
                if sum == S:
                    self.total +=1
            else:
                n = nums[cur]
                find(sum+n,cur+1)
                find(sum-n,cur+1)

        find(0, 0)
        # print(self.total)
        return self.total

assert(Solution().findTargetSumWays([1, 1, 1, 1, 1],3) == 5)   
assert(Solution().findTargetSumWays([7,46,36,49,5,34,25,39,41,38,49,47,17,11,1,41,7,16,23,13],3) == 5756)
```

```
from typing import *

# dp
# TSUM(0:len(nums),S) = TSUM(1:len(nums),S-nums[0]) + TSUM(1:len(nums)ï¼ŒS+nums[0])
class Solution:
    def findTargetSumWays(self, nums: List[int], S: int) -> int:
        mem = {}

        def tsum(i,s):
            if i == len(nums):
                if s == S:
                    return 1
                else:
                    return 0
            else:
                if (i,s) in mem:
                    return mem[(i,s)]
                _tsum = tsum(i+1,s-nums[i]) + tsum(i+1,s+nums[i])
                mem[(i,s)] = _tsum
                return _tsum
        return tsum(0,0)


assert(Solution().findTargetSumWays([1, 1, 1, 1, 1],3) == 5)   
assert(Solution().findTargetSumWays([7,46,36,49,5,34,25,39,41,38,49,47,17,11,1,41,7,16,23,13],3) == 5756)
```
