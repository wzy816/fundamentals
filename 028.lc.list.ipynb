# leetcode 240 Search a 2D Matrix II

Write an efficient algorithm that searches for a value in an m x n matrix. This matrix has the following properties:

- Integers in each row are sorted in ascending from left to right.
- Integers in each column are sorted in ascending from top to bottom.

```
# search from top right to bottom left
# eliminate one row or column at a time
def searchMatrix(matrix, target):
    if not matrix or not matrix[0]:
        return False
    
    rows = len(matrix)
    columns = len(matrix[0])

    
    i = 0
    j = columns-1

    while i<=rows-1 and j>=0:
        v = matrix[i][j]
        if target == v:
            return True
        elif target > v:
            i = i+1
        elif target < v:
            j = j-1
    return False
    
matrix = [
  [1,  4, 7, 11, 15],
  [2,  5, 8, 12, 19],
  [3,  6, 9, 16, 22],
  [10, 13, 14, 17, 24],
  [18, 21, 23, 26, 30]
]            
assert searchMatrix(matrix,5)== True
assert searchMatrix(matrix,20)== False
```

# leetcode 33/81 Search in Rotated Sorted Array

Suppose an array sorted in ascending order is rotated at some pivot unknown to you beforehand.

(i.e., [0,0,1,2,2,5,6] might become [2,5,6,0,0,1,2]).

You are given a target value to search. If found in the array return true, otherwise return false.

```
# use binary search
def search(nums,target):
    print('\nsearch',nums)
    if not nums:
        return False
    
    start = 0
    end = len(nums)-1
    while start <= end:
        print('start=',start,'end=',end)
        center = int((start+end)/2)
        if nums[center] == target:
            return True
        if nums[center]> nums[start]:
            if target >= nums[start] and target <= nums[center]:
                end = center
            else:
                start= center
        elif nums[center] < nums[start]:
            if target >= nums[center] and target <= nums[end]:
                start = center
            else:
                end = center
        else:
            start = start+1
    return False

nums = [2,5,6,0,0,1,2]
assert search(nums,0) == True
assert search(nums,3) == False

nums = [1, 3, 1, 1, 1]
assert search(nums,1) == True
assert search(nums,3) == True
assert search(nums,2) == False

nums = [i for i in range(100,999)]+[i for i in range(0,100)]
assert search(nums,256) == True

nums = [3] + [2 for i in range(100)]
assert search(nums,3) == True
assert search(nums,2) == True
assert search(nums,1) == False

nums = [1,3]
assert search(nums,3) == True
assert search(nums,1) == True
```
