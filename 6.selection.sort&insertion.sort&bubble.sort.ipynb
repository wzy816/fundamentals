```python
"""
Selection Sort
"""
# Selection the smallest item from the remaining unsorted
# And put it at the front
# O(n^2)

def selection_sort(l):
    for i in range(0,len(l)):
        print(i)
        m = l[i]
        m_index = i
        for j in range(i, len(l)):
            if l[j] < m:
                m = l[j]
                m_index = j
        l[i],l[m_index] = l[m_index],l[i]
    return l

# test
l = [7,3,2,5,6,8,9,3]
print(selection_sort(l))
    
```

```python
"""
Insertion Sort
"""
# Insert item into the already sorted
# aka, sorting by insertion
# worse and average O(n^2)
# best O(n)

def insertion_sort(l):
    for i in range(1,len(l)):
        key = l[i]
        j = i-1
        while j >= 0 and l[j]>key:
            l[j+1] = l[j]
            j = j-1
        l[j+1] = key
    return l

# test
l = [7,3,2,5,6,8,9,3]
print(insertion_sort(l))
```

```python
"""
Bubble Sort
"""
# popular but inefficient
# each scan with swap the "smallest" item to the "front"
# so it is safe to ignore the front item from last scan
# aka, sorting by exchange 
# O(n^2)

def bubble_sort(l):
    for i in range(0, len(l)):
        for j in range(len(l)-1,i,-1):
            if l[j-1] > l[j]:
                l[j-1],l[j] = l[j],l[j-1]
    return l

# test
l = [7,3,2,5,6,8,9,3]
print(bubble_sort(l))
```

```python
"""
Shell Sort
"""

# use cirua gap sequence as default
# aka, gapped insertion sort
# https://en.wikipedia.org/wiki/Shellsort
# assume [0:gap-1] is already sorted

def shell_sort(l, gaps = [701,301,132,57,23,10,4,1]):
    for gap in gaps:
        for i in range(gap, len(l)):
            key = l[i]
            j = i-gap
            while j>=0 and l[j]>key:
                l[j+gap] = l[j]
                j = j-gap
            l[j+gap] = key
    return l
    
    
# test
l = [7,3,2,5,6,8,9,3,2,4,5,1,2,7,9,4,3,2,3,5,7,5,4,3,6,7]
print(shell_sort(l))
```
