```python
"""
Insertion Sort
"""
# Insert item into the already sorted
# aka, sorting by insertion
# worse and average O(n^2)
# best O(n)

def insertion_sort(l):
    for i in range(1,len(l)):
        key = l[i]
        j = i-1
        while j >= 0 and l[j]>key:
            l[j+1] = l[j]
            j = j-1
        l[j+1] = key
    return l

# test
l = [7,3,2,5,6,8,9,3]
print(insertion_sort(l))
```
