```python
"""
Bubble Sort
"""
# popular but inefficient
# each scan with swap the "smallest" item to the "front"
# so it is safe to ignore the front item from last scan
# aka, sorting by exchange 
# O(n^2)

def bubble_sort(l):
    for i in range(0, len(l)):
        for j in range(len(l)-1,i,-1):
            if l[j-1] > l[j]:
                l[j-1],l[j] = l[j],l[j-1]
    return l

# test
l = [7,3,2,5,6,8,9,3]
print(bubble_sort(l))
```
