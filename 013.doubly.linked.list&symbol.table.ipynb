```python
"""
Doubly Linked List
"""
class Node():
    def __init__(self,key):
        self._key = key
        self._prev = None
        self._next = None
        
    @property
    def prev(self):
        return self._prev
    
    @property
    def next(self):
        return self._next
    
    @property
    def key(self):
        return self._key
    
    @prev.setter
    def prev(self,prev):
        self._prev = prev
        
    @next.setter
    def next(self,next):
        self._next = next
        
    def __str__(self):
        return '=> ' + repr(self._key)
        
class Nil(Node):
    def __init__(self):
        Node.__init__(self,None)

        
class DoublyLinkedList():
    def __init__(self):
        self._nil = Nil()
        self._nil.prev = self._nil
        self._nil.next = self._nil
        
    def search(self,key):
        x = self._nil.next
        while x != self._nil and x.key != key:
            x = x.next
        return x
    
    def insert(self, x):
        x.next = self._nil.next
        self._nil.next.prev = x
        self._nil.next = x
        x.prev = self._nil
        
    def delete(self, x):
        x.prev.next = x.next
        x.next.prev = x.prev
        
    def display(self):
        ret = ''
        x = self._nil.next
        while x != self._nil:
            print(x)
            x = x.next
        print(ret)
```

```python
dll = DoublyLinkedList()
for i in [1,4,2,5,7]:
    dll.insert(Node(i))
dll.display()

n3 = dll.search(3)
print(n3)
dll.display()

n7 = dll.search(7)
dll.delete(n7)
dll.display()

n2 = dll.search(2)
dll.delete(n2)
dll.display()
```
